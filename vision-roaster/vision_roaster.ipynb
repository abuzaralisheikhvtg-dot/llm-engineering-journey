```python
import discord
import os
import nest_asyncio
import io
import asyncio
import base64
from concurrent.futures import ThreadPoolExecutor
from PIL import Image
from dotenv import load_dotenv
from rich.console import Console
from rich.markdown import Markdown
from rich.panel import Panel
from groq import Groq

console = Console()

# Jupyter support
try:
    from IPython.display import display as ipy_display, Image as IPyImage
    from IPython import get_ipython
    _in_jupyter = get_ipython() is not None
except Exception:
    _in_jupyter = False

# 1. Setup
nest_asyncio.apply()
load_dotenv()

# 2. Groq client
groq_client = Groq(api_key=os.getenv("GROQ_API_KEY"))

# 3. Thread executor
executor = ThreadPoolExecutor(max_workers=3)

# 4. Discord setup
intents = discord.Intents.default()
intents.message_content = True
client = discord.Client(intents=intents)

SYSTEM_PROMPT = """You are a Gen Z brainrot meme lord who roasts photos like an unhinged Instagram comment section.

Structure your roast EXACTLY like this:

**first impression:** act like you just scrolled past this on your feed and physically recoiled. 3-4 sentences. use lowercase. be chaotic.

**ratio + L:** this is where you deliver the most unhinged comparison. what does this person look like? what NPC are they? what bootleg version of what character? 3-4 sentences.

**final boss roast:** end with the most sigma brain closing statement. make it quotable. make it hurt. 2-3 sentences.

RULES:
- write in full lowercase unless something is SO bad it needs CAPS
- use gen z brainrot vocabulary: npc, ratio, L, W, slay, no cap, fr fr, bussin, mid, rizz, sigma, ohio, skibidi, delulu, ate and left no crumbs, it's giving, main character, side quest, caught in 4k
- use emojis AGGRESSIVELY 💀☠️😭🙏💅🗿😤
- short punchy sentences. NO sophisticated words. if a word has more than 3 syllables replace it with "idk some big word"
- minimum 200 words no cap fr fr
- roast ANYTHING in the photo — food, objects, people, animals, a wall, literally anything
- be unhinged. be chaotic. be brainrot."""


def startup_test():
    console.print("[cyan]🔍 Testing Groq API connection...[/cyan]")
    try:
        r = groq_client.chat.completions.create(
            model="meta-llama/llama-4-scout-17b-16e-instruct",
            max_tokens=20,
            messages=[{"role": "user", "content": "Say: API OK"}]
        )
        console.print(f"[green]✅ Groq API working: {r.choices[0].message.content.strip()}[/green]")
        return True
    except Exception as e:
        console.print(f"[bold red]❌ Groq API FAILED: {e}[/bold red]")
        return False


def call_groq(img_bytes: bytes) -> str:
    img_b64 = base64.standard_b64encode(img_bytes).decode("utf-8")
    response = groq_client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",
        max_tokens=1024,
        messages=[
            {
                "role": "system",
                "content": SYSTEM_PROMPT
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/jpeg;base64,{img_b64}"
                        }
                    },
                    {
                        "type": "text",
                        "text": "roast this image. go absolutely feral. no mercy 💀"
                    }
                ]
            }
        ]
    )
    return response.choices[0].message.content


def show_in_terminal(roast_text):
    console.print(Panel(
        Markdown(roast_text),
        title="🔥 [bold red]ROAST[/bold red]",
        border_style="red",
        expand=False
    ))


def show_in_notebook(img, roast_text):
    if not _in_jupyter:
        return
    try:
        from IPython.display import display as d, Markdown as MD
        buf = io.BytesIO()
        img.save(buf, format="PNG")
        buf.seek(0)
        d(IPyImage(data=buf.getvalue()))
        d(MD(f"**🔥 Roast:**\n\n{roast_text}"))
    except Exception as e:
        console.print(f"[yellow]Notebook display failed: {e}[/yellow]")


@client.event
async def on_ready():
    console.print(f"[bold green]--- VisionRoaster LIVE as {client.user} ---[/bold green]")


@client.event
async def on_message(message):
    if message.author == client.user:
        return

    if message.content.lower() == "ping":
        await message.reply("Pong! Send an image if you dare. 💀")
        return

    if not message.attachments:
        return

    for attachment in message.attachments:
        if not any(attachment.filename.lower().endswith(ext)
                   for ext in ['png', 'jpg', 'jpeg', 'webp', 'gif']):
            continue

        console.print(f"[yellow]📥 Image from {message.author} — roasting...[/yellow]")

        async with message.channel.typing():
            try:
                # Step A: Download + resize
                img_data = await attachment.read()
                img = Image.open(io.BytesIO(img_data)).convert("RGB")
                img.thumbnail((1024, 1024))

                # Prepare bytes
                buf = io.BytesIO()
                img.save(buf, format="JPEG", quality=85)
                img_bytes = buf.getvalue()

                console.print(f"[cyan]📤 Sending to Groq ({len(img_bytes)//1024}KB)...[/cyan]")

                # Step B: Call Groq in executor thread
                loop = asyncio.get_event_loop()
                roast_text = await asyncio.wait_for(
                    loop.run_in_executor(executor, call_groq, img_bytes),
                    timeout=30
                )

                console.print(f"[green]✅ Roast ready — {len(roast_text)} chars[/green]")

                # Step C: Display in terminal + notebook
                show_in_terminal(roast_text)
                if _in_jupyter:
                    show_in_notebook(img, roast_text)

                # Step D: Send to Discord in chunks
                chunks = [roast_text[i:i+1990]
                          for i in range(0, len(roast_text), 1990)]
                for chunk in chunks:
                    await message.reply(chunk)

                console.print(f"[green]📤 Sent {len(chunks)} chunk(s) to Discord[/green]")

            except asyncio.TimeoutError:
                console.print("[red]❌ Timed out after 30s[/red]")
                await message.reply("⏱️ timed out fr fr try again 😭")
            except Exception as e:
                console.print(f"[bold red]❌ Error: {type(e).__name__}: {e}[/bold red]")
                await message.reply(f"something broke no cap: {str(e)[:200]}")


# 5. Run
if startup_test():
    client.run(os.getenv("DISCORD_TOKEN"))
else:
    console.print("[bold red]Fix your GROQ_API_KEY before starting.[/bold red]")
```
